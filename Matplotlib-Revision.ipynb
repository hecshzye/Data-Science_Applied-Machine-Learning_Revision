```python
# Making plots from a dictionary

nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Hecshzye's Nut Butter Store",
      ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()

#barh() for horizontal

ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

### Multiple plots (Subplots)

#### type 1 syntax

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(12, 8))

# plotting to each different azis

ax1.plot(x, x/2);
ax2.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax3.scatter(np.random.random(10), np.random.random(10));
ax4.hist(np.random.randn(2800));
```

#### type 2 syntax

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(12, 8))

# one each different index

ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

```python
# new df

heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease["age"].plot.hist(bins=60);
```

```python
heart_disease.head(10)
```

```python
over_60 = heart_disease[heart_disease["age"] > 60]
over_60.head(10)
```

```python
# pyplot viz

over_60.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            cmap="winter",
            figsize=(12, 7));
```

```python
# using matplotlib (oo method)

fix, ax = plt.subplots(figsize=(10, 7))
over_60.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             cmap="viridis",
             ax=ax);

ax.set_xlim([45, 100]);
```

```python
# object oriented method 

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_60["age"],
                     y=over_60["chol"],
                     c=over_60["target"],
                     cmap="plasma");

ax.set(title="Heart Disease Analysis",
       xlabel="Age",
       ylabel="Cholesterol");
      
ax.legend(*scatter.legend_elements(), title="Target");

ax.axhline(over_60["chol"].mean(),
          linestyle='--');
ax.axvline(over_60["age"].mean(),
          linestyle='--');
```

```python
# subplotting 'chol', 'age', 'thalach' (heartrate)

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(12,12),
                               sharex=True)

scatter = ax0.scatter(x=over_60["age"],
                      y=over_60["chol"],
                      c=over_60["target"])
ax0.set(title="Heart Disease Analysis",
        ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(y=over_60["chol"].mean(),
            linestyle="--");

#ax1

scatter = ax1.scatter(x=over_60["age"],
                      y=over_60["thalach"],
                      c=over_60["target"])
ax1.set(title="Heart Disease Analysis and Maximum Heartrate",
        xlabel="Age",
        ylabel="Maximum HeartRate")
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.axhline(y=over_60["thalach"].mean(),
            linestyle="--");

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
